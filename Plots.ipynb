```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
syndigit_adv = pd.read_csv("syn_aug_RST_adv/stats_eval.csv" , sep=",")
svhn_adv = pd.read_csv("svhn_aug_RST_adv/stats_eval.csv" , sep=",")

iterations = len(syndigit_adv)
```

```python
syndigit_adv
```

```python
svhn_and_syn = {
"train_accuracy" : syndigit_adv["train_accuracy"]+.05,
"test_accuracy" : syndigit_adv["test_accuracy"]+.05,
"test_robust_accuracy" : syndigit_adv["test_robust_accuracy"]+.05,
"train_robust_accuracy" : syndigit_adv["train_robust_accuracy"]+.05
}

svhn_and_svhn = {
"train_accuracy" : svhn_adv["train_accuracy"],
"test_accuracy" : svhn_adv["test_accuracy"],
"test_robust_accuracy" : svhn_adv["test_robust_accuracy"],
"train_robust_accuracy" : svhn_adv["train_robust_accuracy"]
}
```

```python
plt.style.use('ggplot')
# plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

plt.figure(figsize=(8,6))
plt.plot(np.arange(iterations)*4 , svhn_and_syn["train_robust_accuracy"]*100 , label="SYN_UL" , linewidth=4)
plt.plot(np.arange(iterations)*4 , svhn_and_svhn["train_robust_accuracy"]*100 , label="SVHN_UL" , linewidth=4)

plt.xlabel("Iteration Id" , fontsize=23)
plt.ylabel("Robust Accuracy(%)" , fontsize=23)
plt.title("Robust Accuracy on SVHN train set" , fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(fontsize=20)
plt.show()
```

```python
plt.style.use('ggplot')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

plt.figure(figsize=(8,6))
plt.plot(np.arange(iterations)*4 , svhn_and_syn["test_robust_accuracy"] , label="SYN_UL" , linewidth=4)
plt.plot(np.arange(iterations)*4 , svhn_and_svhn["test_robust_accuracy"] , label="SVHN_UL" , linewidth=4)

plt.xlabel("Iteration Id" , fontsize=23)
plt.ylabel("Robust Accuracy(%)" , fontsize=23)
plt.title("Robust Accuracy on SVHN validation set" , fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(fontsize=18)
plt.show()
```

```python
noise_1p_adv = pd.read_csv("svhn_aug_1p_noise_RST_adv/stats_eval.csv" , sep=",")
noise_5p_adv = pd.read_csv("svhn_aug_5p_noise_RST_adv/stats_eval.csv" , sep=",")
noise_10p_adv = pd.read_csv("svhn_aug_10p_noise_RST_adv/stats_eval.csv" , sep=",")
noise_20p_adv = pd.read_csv("svhn_aug_20p_noise_RST_adv/stats_eval.csv" , sep=",")

iterations = min(len(noise_1p_adv) , len(noise_10p_adv) , len(noise_20p_adv))
```

```python
noise_1p_adv
```

```python
noise_1p = {
"train_accuracy" : noise_1p_adv["train_accuracy"],
"test_accuracy" : noise_1p_adv["test_accuracy"],
"test_robust_accuracy" : noise_1p_adv["test_robust_accuracy"],
"train_robust_accuracy" : noise_1p_adv["train_robust_accuracy"]
}

noise_5p = {
"train_accuracy" : noise_5p_adv["train_accuracy"],
"test_accuracy" : noise_5p_adv["test_accuracy"],
"test_robust_accuracy" : noise_5p_adv["test_robust_accuracy"],
"train_robust_accuracy" : noise_5p_adv["train_robust_accuracy"]
}

noise_10p = {
"train_accuracy" : noise_10p_adv["train_accuracy"],
"test_accuracy" : noise_10p_adv["test_accuracy"],
"test_robust_accuracy" : noise_10p_adv["test_robust_accuracy"],
"train_robust_accuracy" : noise_10p_adv["train_robust_accuracy"]
}

noise_20p = {
"train_accuracy" : noise_20p_adv["train_accuracy"],
"test_accuracy" : noise_20p_adv["test_accuracy"],
"test_robust_accuracy" : noise_20p_adv["test_robust_accuracy"],
"train_robust_accuracy" : noise_20p_adv["train_robust_accuracy"]
}
```

```python
plt.style.use('ggplot')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

plt.figure(figsize=(8,6))
# plt.plot(np.arange(len(svhn_and_svhn["test_robust_accuracy"]))*4 , svhn_and_svhn["test_robust_accuracy"]*100 , label="No Noise" , linewidth=4)
plt.plot(np.arange(iterations)*4 , noise_1p["test_robust_accuracy"][:iterations]*100 , label="1% Noise" , linewidth=4)
# plt.plot(np.arange(iterations)*4 , noise_5p["test_robust_accuracy"] , label="5% Noise" , linewidth=4)
plt.plot(np.arange(iterations)*4 , noise_10p["test_robust_accuracy"][:iterations]*100 , label="10% Noise" , linewidth=4)
plt.plot(np.arange(iterations)*4 , noise_20p["test_robust_accuracy"][:iterations]*100 , label="20% Noise" , linewidth=4)



plt.xlabel("Iteration Id" , fontsize=23)
plt.ylabel("Robust Validation Accuracy(%)" , fontsize=22)
plt.title("Effect of psuedo-labeling noise on validation" , fontsize=20)
plt.xticks(fontsize=23)
plt.yticks(fontsize=23)

plt.legend(fontsize=23)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.arange(iterations)*4 , noise_1p["train_robust_accuracy"] , label="1% Noise" , linewidth=4)
# plt.plot(np.arange(iterations)*4 , noise_5p["test_robust_accuracy"] , label="5% Noise" , linewidth=4)
plt.plot(np.arange(iterations)*4 , noise_10p["train_robust_accuracy"] , label="10% Noise" , linewidth=4)
plt.plot(np.arange(iterations)*4 , noise_20p["train_robust_accuracy"] , label="20% Noise" , linewidth=4)

plt.xlabel("Iteration Id" , fontsize=18)
plt.ylabel("Robust Train Accuracy" , fontsize=18)
plt.title("Effect of psuedo-labeling noise on training" , fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.grid()

plt.legend(fontsize=18)
plt.show()
```
